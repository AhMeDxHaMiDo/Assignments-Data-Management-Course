```python
import pandas as pd
```

```python
try:
    import pymongo
    print("Module Import Successful")
except ImportError as error:
    print("Module Import Error")
    print(error)
```

```python
city = pd.read_csv('city.csv', sep=';')
event = pd.read_csv('event.csv', sep=';')
games = pd.read_csv('games.csv', sep=';')
games_city = pd.read_csv('games_city.csv', sep=';')
competitor_event = pd.read_csv('competitor_event.csv')
games_competitor = pd.read_csv('games_competitor.csv', sep=';')
person = pd.read_csv('person.csv', sep=';')
medal = pd.read_csv('medal.csv', sep=';')
person_region = pd.read_csv('person_region.csv', sep=';')
sport = pd.read_csv('sport.csv', sep=';')
```

```python
city_gamescity = pd.merge(city, games_city, how='inner', 
                          left_on=['id'], right_on=['city_id']).drop(columns=['id', 
                                                                              'city_id', 
                                                                              'country_id'])

games_final = pd.merge(games, city_gamescity, how='inner', 
                            left_on=['id'], right_on=['games_id']).drop(columns=['id'])
games_final.head()
```

```python
person.columns, person_region.columns
```

```python
personregion_person = pd.merge(person, person_region, how='inner', 
                               left_on=['id'], right_on=['person_id']).drop(columns=['id', 
                                                                                     'region_id'])
personregion_person.head()
```

```python
person_final = pd.merge(games_competitor, personregion_person, 
                        how='inner', on='person_id').drop(columns=['id', 
                                                                   'games_id'])
person_final.head()
```

```python
games_competitor.drop(columns='age', inplace=True)
```

```python
medal_competitorevent = pd.merge(medal, competitor_event, how='inner', 
                                 left_on=['id'], right_on=['medal_id']).drop(columns=['id', 
                                                                                      'medal_id'])
medal_competitorevent.head()
```

```python
event_sport = pd.merge(event, sport, how='inner', 
                       left_on=['sport_id'], right_on=['id'],
                       suffixes=('', '_sport')).drop(columns=['sport_id', 
                                                              'id_sport'])
event_sport
```

```python
person_final.columns
```

```python
competitorevent_final = pd.merge(medal_competitorevent, event_sport, how='inner', 
                                      left_on=['event_id'], right_on=['id']).drop(columns=['event_id', 
                                                                                           'id'])
competitorevent_final.head()
```

```python
games_competitor_final = pd.merge(competitorevent_final, games_competitor, how='inner',
                                  left_on=['competitor_id'], right_on=['id']).drop(columns=['competitor_id', 'id'])
games_competitor_final.head()
```

```python
games_final.index = games_final.index.map(str)
```

```python
games_final
```

```python
games_metadata = games_final.to_dict('records')
```

```python
games_metadata
```

```python
person_games = pd.merge(games_competitor_final, person_final, 
                        how='inner', on=['person_id']).drop(columns=['person_id'])
display(person_games.shape)
person_games = person_games.to_dict('records')
```

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["mydatabase"]
mycol = mydb["games_metadata"]

x = mycol.insert_many(games_metadata)

#print list of the _id values of the inserted documents:
print(x.inserted_ids)
```
